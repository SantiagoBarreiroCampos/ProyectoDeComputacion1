```python
import pandas as pd
import os
import numpy as np
```

```python
odioPath = os.getcwd() + "\\Odio"
odioPath
```

```python
files = os.listdir(odioPath)
files
```

```python
odioArray = []
for file in files:
        file_path = os.path.join(odioPath, file)

        
        if os.path.isfile(file_path):
            with open(file_path, 'r', encoding="ISO 8859-1") as f:
                
                fileread = f.read();
                splitedcontent = fileread.split("\n#####\n")#0=link 1=autor 2=fecha 3=titulo 4=content
                #juntar title y content
                content=splitedcontent[3]+" "+splitedcontent[4]+" "+splitedcontent[5]
                odioArray.append(content)
                
                
type(odioArray[0])
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.snowball import SnowballStemmer

def noticiasLower(arrayNoticias):
    noticiaslower = []
    for noticia in arrayNoticias:
        noticiaslower.append(noticia.lower())
    return noticiaslower

def tokenize(arrayNoticias):
    noticiasTokenizadas = []
    tokenizer = RegexpTokenizer(r"[a-zA-ZáéíóúñÁÉÍÓÚÑ]+")
    for noticia in arrayNoticias:
        noticiasTokenizadas.append(tokenizer.tokenize(noticia))
    
    return noticiasTokenizadas

def removeStopwords(noticia):
    with open("listaParadaEsp.txt") as f:
        text = f.read()
        prohibitedWords = text.split("\n")
        return [word for word in noticia if not word in prohibitedWords]
    
def stemming(x):
    stemmer = SnowballStemmer(language="spanish")
    return ' '.join([stemmer.stem(word)for word in x])
```

```python
noticiasMinus = []
noticiasMinus = noticiasLower(odioArray)

noticiasTokenizadas = []
noticiasTokenizadas = tokenize(noticiasMinus)

#noticiasTokenizadas[0]
```

```python
noticiasSinStop = []

for noticia in noticiasTokenizadas:
    noticiasSinStop.append(removeStopwords(noticia))

#noticiasSinStop[0]
```

```python
noticiasStem = []

for noticia in noticiasSinStop:
    noticiasStem.append(stemming(noticia))

#noticiasStem[6]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(noticiasStem)
palabras = vectorizer.get_feature_names()
array = X.toarray()
#nombres
```

```python
tabla = pd.DataFrame(data = array, index = files, columns = palabras)
tabla
```
